```python
from google.colab import files
import io
import pandas as pd

uploaded = files.upload()
fname = list(uploaded.keys())[0]

df = pd.read_csv(io.BytesIO(uploaded[fname]), sep=';')
df.head()
```

```python
# Proses lengkap pada pengelolaan data hingga data evaluation menggunakan open dataset(dari kaggle)
# Import Library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# DATA UNDERSTANDING
print("\n=== INFORMASI DATA ===")
df.info()

print("\n=== CEK NILAI HILANG ===")
print(df.isnull().sum())

print("\n=== DESKRIPSI DATA ===")
display(df.describe())

# DATA CLEANING
# Hapus duplikat
df = df.drop_duplicates()

# Isi nilai kosong dengan median
df = df.fillna(df.median(numeric_only=True))

print("\n=== SETELAH CLEANING ===")
display(df.head())

# DETEKSI & PENANGANAN OUTLIER (METODE IQR)
for col in df.select_dtypes(include=np.number).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR

    # Hapus outlier
    df = df[(df[col] >= batas_bawah) & (df[col] <= batas_atas)]

print("\n=== DATASET SETELAH OUTLIER DIHAPUS ===")
display(df.head())

# Visualisasi outlier (contoh kolom pertama)
num_col = df.select_dtypes(include=np.number).columns[0]
plt.figure(figsize=(8,4))
sns.boxplot(x=df[num_col])
plt.title(f"Boxplot {num_col} Setelah Penanganan Outlier")
plt.show()

# DATA PREPROCESSING (NORMALISASI)
scaler = MinMaxScaler()
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = scaler.fit_transform(df[num_cols])

print("\n=== DATASET SETELAH NORMALISASI ===")
display(df.head())

# VISUALISASI DATA
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Korelasi Antar Variabel Numerik")
plt.show()

# Histogram distribusi data
df.hist(figsize=(12,8), bins=20)
plt.suptitle("Distribusi Data Setelah Normalisasi")
plt.show()

# EVALUASI DATA
print("\n=== EVALUASI DATA ===")
display(df.describe())

# Korelasi tertinggi antar variabel
corr = df.corr()
print("\nKorelasi tertinggi antar variabel:")
print(corr.unstack().sort_values(ascending=False).drop_duplicates()[1:6])

# Info tambahan evaluasi
print("\nJumlah data akhir setelah cleaning & outlier removal:", df.shape[0])
print("Jumlah fitur (kolom):", df.shape[1])

# SIMPAN DATA AKHIR
df.to_csv('data_cleaned_final.csv', index=False)
print("\nâœ… Data akhir telah disimpan sebagai 'data_cleaned_final.csv'")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Membaca dataset cuaca
data = pd.read_csv('seattle-weather.csv')
print("5 data teratas:")
print(data.head())
print("\nInfo dataset:")
print(data.info())
```

```python
# Hapus kolom 'date' karena bukan fitur numerik
data = data.drop(columns=['date'])

# Cek nilai kosong
print("\nCek missing values:")
print(data.isnull().sum())

# Pisahkan fitur dan label
X = data[['precipitation', 'temp_max', 'temp_min', 'wind']]
y = data['weather']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
model = GaussianNB()
model.fit(X_train, y_train)

# Prediksi
y_pred = model.predict(X_test)

print("Akurasi Model:", accuracy_score(y_test, y_pred))
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
sns.heatmap(cm, annot=True, fmt='d', cmap='YlOrBr',
            xticklabels=model.classes_, yticklabels=model.classes_)
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.title('Confusion Matrix - Seattle Weather Classification')
plt.show()
```

```python
# Menampilkan rata-rata fitur tiap kelas (hasil training GaussianNB)
mean_df = pd.DataFrame(model.theta_, columns=X.columns, index=model.classes_)
print("\nRata-rata fitur per kelas:")
print(mean_df)
```

```python
cm_df = pd.DataFrame(cm, index=model.classes_, columns=model.classes_)
print("\nConfusion Matrix dalam bentuk tabel:")
print(cm_df)
```
